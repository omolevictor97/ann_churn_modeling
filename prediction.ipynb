```python
import tensorflow as tf
from tensorflow.keras.models import load_model
import joblib
import pandas as pd
import numpy as np
```

```python
# load all necessary files
model_tf = load_model("model_tf.h5")

loaded_preprocessor = joblib.load("preprocessor.joblib")
```

```python
# Example data for prediction
input_data = {
    "CreditScore" : 600,
    "Geography" : "France",
    "Gender" : "male".capitalize(),
    "Age" : 40,
    "Tenure" : 3,
    "Balance" : 60000,
    "NumOfProducts" : 2,
    "HasCrCard" : 1,
    "IsActiveMember" : 1,
    "EstimatedSalary" : 50000
}
input_df = pd.DataFrame(data=[input_data])
preprocessed_arr = loaded_preprocessor.transform(input_df)
preprocessed_arr
```

```python
0 if model_tf.predict(preprocessed_arr)[0][0] < 0.5 else 1
```
